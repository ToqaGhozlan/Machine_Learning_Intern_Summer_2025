```python
def remove_outliers_strict(df, column='fare_amount', group='Weather'):
    def filter_group(g):
        Q1 = g[column].quantile(0.25)
        Q3 = g[column].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        return g[(g[column] >= lower) & (g[column] <= upper)]

    return df.groupby(group, group_keys=False).apply(filter_group)

df_clean = remove_outliers_strict(df)

sns.boxplot(x='Weather', y='fare_amount', data=df_clean)
plt.title('Fare vs Weather (Outliers Removed)')
plt.show()
```

```python
avg_fares = df.groupby('Weather')['fare_amount'].mean().reset_index()

avg_fares = avg_fares.sort_values(by='fare_amount', ascending=False)

sns.barplot(x='Weather', y='fare_amount', data=avg_fares, palette='viridis')
plt.title('Average Fare by Weather')
plt.ylabel('Average Fare Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_traffic = df.groupby('Traffic Condition')['fare_amount'].mean().reset_index()
avg_traffic = avg_traffic.sort_values(by='fare_amount', ascending=False)

sns.barplot(x='Traffic Condition', y='fare_amount', data=avg_traffic, palette='viridis')
plt.title('Average Fare by Traffic Condition')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_day = df.groupby('day_of_week')['fare_amount'].mean().reset_index()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
avg_day['day_of_week'] = pd.Categorical(avg_day['day_of_week'], categories=day_order, ordered=True)
avg_day = avg_day.sort_values('day_of_week')

sns.barplot(x='day_of_week', y='fare_amount', data=avg_day, palette='viridis')
plt.title('Average Fare by Day of Week')
plt.tight_layout()
plt.show()
```
