```python
df.drop(["P-C","P-not-C","date of reservation"],axis=1,inplace=True)
```

```python
X = df.drop('booking status', axis=1)
y = df['booking status']
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
X_encoded = X.copy()
categorical_cols = ['type of meal', 'room type', 'market segment type']
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)



```

```python
le_y = LabelEncoder()
y_encoded = le_y.fit_transform(y)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
import pandas as pd

selector = SelectKBest(score_func=f_classif, k=5)
X_new = selector.fit_transform(X_encoded, y_encoded)


dfscores = pd.DataFrame(selector.scores_, columns=["Score"])
dfcolumns = pd.DataFrame(X_encoded.columns, columns=["Feature"])


featureScores = pd.concat([dfcolumns, dfscores], axis=1)
print(featureScores.sort_values(by='Score', ascending=False))
```

```python
from sklearn.feature_selection import SelectKBest, chi2

bestfeatures = SelectKBest(score_func=chi2, k=5)
fit = bestfeatures.fit(X_encoded, y_encoded)
dfscores = pd.DataFrame(selector.scores_, columns=["Score"])
dfcolumns = pd.DataFrame(X_encoded.columns, columns=["Feature"])


featureScores = pd.concat([dfcolumns, dfscores], axis=1)
print(featureScores.sort_values(by='Score', ascending=False))
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif
selector = SelectKBest(score_func=mutual_info_classif, k=5)
X_new = selector.fit_transform(X_encoded, y_encoded)
featureScores = pd.concat([dfcolumns, dfscores], axis=1)
print(featureScores.sort_values(by='Score', ascending=False))
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

model = RandomForestClassifier()
model.fit(X_encoded, y_encoded)

importances = model.feature_importances_
features = X_encoded.columns

feature_importance = pd.DataFrame({
    'Feature': features,
    'Importance': importances
})

feature_importance.sort_values(by='Importance', ascending=False, inplace=True)

print(feature_importance)
```

```python
X_encoded = X_encoded.drop(
    ["type of meal_Meal Plan 3", "type of meal_Not Selected", "market segment type_Corporate",
     "room type_Room_Type 3", "room type_Room_Type 2"],
    axis=1
)

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

model = RandomForestClassifier()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=["Canceled", "Not Canceled"]))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(
    y_test, y_pred,
    display_labels=["Canceled", "Not Canceled"],
    cmap='Reds'
)
plt.title("Confusion Matrix")
plt.show()
```
