```python
def detect_outliers_zscore():
    res = []
    nums = []
    for column in newdf.columns:
        if newdf[column].dtype in ['int64', 'Int64', 'float64']:
            col_mean = newdf[column].mean()
            col_std = newdf[column].std()
        
            z_scores = (newdf[column] - col_mean) / col_std
            if (np.abs(z_scores) > 3).sum() > 0:
                res.append(column) 
                nums.append((np.abs(z_scores) > 3).sum())
            
    return (res,nums)
```

```python
outliers = detect_outliers_zscore()
outliers
```

```python
newdf = newdf.apply(lambda col: col.fillna(col.median()) if pd.api.types.is_numeric_dtype(col) else col)
```

```python
from sklearn.preprocessing import LabelEncoder

def transform_categorical_features():
    global newdf
    newdf = newdf.copy()
    label_encoder = LabelEncoder()

    for col in newdf.select_dtypes(include='object').columns:
        unique_vals = newdf[col].nunique()

        if unique_vals == 2:
            # Label encode binary columns
            newdf[col] = label_encoder.fit_transform(newdf[col])
        else:
            # One-hot encode multiclass columns
            dummies = pd.get_dummies(newdf[col], prefix=col, drop_first=True)
            newdf = pd.concat([newdf.drop(columns=[col]), dummies], axis=1)

    return newdf
```

```python
newdf = transform_categorical_features()
```

```python
newdf
```

```python
X = newdf.drop(['booking status', 'date of reservation'], axis=1)
y = newdf['booking status']
```

```python
X.isna().sum()
```

```python
X = X.astype(np.float64)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state=42)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
cr = classification_report(y_test, y_pred)
print(cr)
```
